# Dask arrays replace numpy arrays

With numpy arrays, we can perform many numpy array features in parallel.

Dask arrays are collections of numpy arrays. They are similar to numpy arrays, but their chunks are stored in different places.  The user specifies the size of the chunks.  How to decide the chunk size?

"In general, a chunk size between 100MB and 1GB is good," but the best value depends on your data, your cluster, and what you are trying to do.  You may want to try a few different values to see what works best for you.

```python
# Create a random Dask array
x = da.random.random((4000, 4000), chunks=(2000, 2000))
x
```

```python
# Perform a computation
y = x.mean()
y.visualize()
```

Let's go a bit bigger to showcase the speed advantage of processing in parallel.

```python
x = da.random.random((40000, 40000), chunks=(2000, 2000))
y = x.mean()
```

Much like dask delayed, this does not actually perform the computation, it just sets up the computation.  We need to call compute to actually perform the computation.

```python
%%time
y.compute()
```

Here is the same in traditional numpy.

```python
%%time
x = np.random.random((40000, 40000))
x.mean()
```

## Simple math operations

```python
x = da.random.random(1_000_000, chunks=200_000)
y = da.random.random(1_000_000, chunks=200_000)
```

```python
z = da.sqrt(x ** 2 + y ** 2)
```

```python
z.visualize()
```

```python
%%time
z.compute()
```

## Assigning and indexing

```python
z_trimmed = z.copy()
z_trimmed[z_trimmed < .2] = 0
```

```python
df = pd.DataFrame({'z':z.compute(), 'z_trimmed':z_trimmed.compute()})
```

```python
px.histogram(df, x=['z', 'z_trimmed'], range_y=[0,10_000], barmode='overlay')
```

## Shortcomings

Chunking and aggregating assumes some homogeneity in the data when dealing with sorting/ordering operations like sort, percentile etc, and in these cases dask array is not the best choice.

```python
x = da.random.random(1_000_000, chunks=200_000)
y = da.percentile(x, 50)
```

```python
y.visualize()
```

See the problem?  If our data is non-uniform in some way, the highest values might all be in one chunk.

```python
x = np.random.normal(size=1_000_000)
x = np.sort(x)
y = np.percentile(x, 99)
y
```

```python
dask_x = da.from_array(x, chunks=200_000)
dask_y = da.percentile(dask_x, 99)
print(dask_y.compute())

dask_x = da.from_array(x, chunks=100_000)
dask_y = da.percentile(dask_x, 99)
print(dask_y.compute())

dask_x = da.from_array(x, chunks=50_000)
dask_y = da.percentile(dask_x, 99)
print(dask_y.compute())
```

Different chunk sizes give different results, and none of them are correct.

# Image processing

Dask array has two powerful functions for image processing: map_blocks and map_overlap.

map_blocks is a parallel version of map.  It applies a function to each chunk of the array, and returns a new dask array, it can take in any function that operates on numpy arrays.

map_overlap is the more useful one, as it windows the chunks and applies a function to the overlapping regions.  This is useful for gaussian blurs, edge detection, and other functions that require information from neighboring pixels.

```python
x = np.random.random(size=(4000, 4000))
peak_positions = np.random.randint(0, 4000, size=(2000, 2))
x[peak_positions[:,0], peak_positions[:,1]] = 100
x = filters.gaussian(x, sigma=10)
```

```python
plt.figure(figsize=[10,10])
plt.imshow(x, vmin=0, vmax=1)
```

Let's apply a traditional peak accentuator:  Laplacian of gaussian in numpy.

```python
%%time
plt.figure(figsize=[10,10])
y = -ndi.gaussian_laplace(x, sigma=10)
plt.imshow(y, vmin=-.0006, vmax=.002)
```

Now let's do it in dask arrays

```python
%%time
plt.figure(figsize=[10,10])
dask_x = da.from_array(x, chunks=(2000, 2000))
dask_y = -da.map_overlap(ndi.gaussian_laplace, dask_x, sigma=10, depth=20, boundary='reflect')
plt.imshow(dask_y.compute(), vmin=-.0006, vmax=.002)
```

What happens if we change the "depth" parameter

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[20,10])
dask_x = da.from_array(x, chunks=(2000, 2000))
dask_y = -da.map_overlap(ndi.gaussian_laplace, dask_x, sigma=10, depth=20, boundary='reflect')
bad_dask_y = -da.map_overlap(ndi.gaussian_laplace, dask_x, sigma=10, depth=0, boundary='reflect')


ax1.imshow(bad_dask_y.compute()[1500:2500, 1500:2500])
ax1.set_title('Depth = 0, no padding of windowed chunks')
ax2.imshow(dask_y.compute()[1500:2500, 1500:2500])
ax2.set_title('Depth = 20, padding windowed chunks 20 pixels on each side')
```

## Processing very large images

```python
%%time
# Create a random Dask array
x = da.random.random((40, 40000, 40000), chunks=(1000, 2000, 2000))
print(x)

# Perform a computation
y = x.mean()

# Compute the result
result = y.compute()
print("Mean:", result)
```

# Lazy Loading

Dask has the option of loading arrays lazily, ie leaving them on disk and only accessing when necessary.  Napari can take advantage of this when displaying images so it need not all be in memory (similar to virtual stack in imagej)

```python
import napari
```

```python
viewer = napari.Viewer()
```

```python
fnames = glob.glob('tiffs/*.tif')
```

We load the first one just to get the dimensions

```python
sample = tifffile.imread(fnames[0])
```

We first setup a lazy reader using dask delayed, then we create a dask array from the delayed reader.

```python
lazy_imread = dask.delayed(tifffile.imread)  # lazy reader
lazy_arrays = [lazy_imread(fn) for fn in fnames]
dask_arrays = [da.from_delayed(delayed_reader, shape=sample.shape, dtype=sample.dtype) for delayed_reader in lazy_arrays]
# Stack into one large dask.array
stack = da.stack(dask_arrays, axis=0)
stack.shape  # (nfiles, ny, nx)

stack
```

This is not very good dasking, so we will rechunk it

```python
stack = stack.rechunk(chunks=(4, 512, 512))
stack
```

Now we will perform some operations on the array, just as if it was a normal image

```python
smoothed = da.map_overlap(skimage.filters.gaussian, stack, depth=10, sigma=20, preserve_range=True, boundary='reflect')
back_sub = stack - 0.7*smoothed
da.clip(back_sub, 0, 655356, out=back_sub)
```

Add them to napari, and scroll through in Z:  this is not actually in memory, it is just being loaded as we scroll through.

```python
viewer.add_image(stack, name='raw, lazy')
viewer.add_image(smoothed, name='smoothed, lazy')
viewer.add_image(back_sub, name='back_sub, lazy')
```

Scrolling through the different layers responds differently.  "smoothed" and "back_sub" are lazy loaded AND lazy processed, so they are only loaded and processed when we scroll through them.  "raw" is lazy loaded, but not lazy processed, so it is loaded, but then processed in memory.  If you watch CPU usage when scrolling through 'raw' you will see no usage, (even though it is loading from disk as it updates), but when you scroll through 'smoothed' or 'back_sub' you will see CPU usage max out the first time you scroll through a slice and then it will be cached in memory.

```python
viewer.add_image(back_sub.compute(), name='back_sub, in memory')
```

A computed version on other hand is loaded fully into memory and scrolls more rapidly and requires no CPU.
